```python
#PCK DEP
import requests
from pyquery import PyQuery 
```

```python
url = 'https://github.com/globalcitizen/2019-wuhan-coronavirus-data/tree/master/data-sources/bno/data'
urlraw = 'https://raw.githubusercontent.com/globalcitizen/2019-wuhan-coronavirus-data/master/data-sources/bno/data'
response = requests.get(url)
#print(response.text)
   
html = response.text
pq = PyQuery(html)
#tag = pq('a') # or     tag = pq('div.class')
#print (tag)
#print(tag.attr("title"))
#class="js-navigation-open"
filtered = pq('a').filter(lambda i: pq(this).text().endswith('.csv'))
#print(filtered.text())
listFiles = filtered.text().split(" ")
```

```python
#Get content data
datatotal =''
for item in listFiles:
    urlcsv = urlraw + "/" + item
    response = requests.get(urlcsv)
    data = response.text
    #print(urlcsv)
    #print(data)
    datatotal = datatotal + data
    #pq = PyQuery(html)
    #filtered = pq('a').filter(lambda i: pq(this).text().endswith('.csv')
print(datatotal)
```
